```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from function import *
from tqdm import tqdm

warnings.simplefilter('ignore')
```

```python
data_path = '../1.data/'
train_file_name = 'train.csv'
test_file_name  = 'test.csv'
```

```python
train_df = pd.read_csv(f'{data_path}{train_file_name}')
test_df = pd.read_csv(f'{data_path}{test_file_name}')
train_df.head()
```

```python
non_cols = ['id', 'price']
num_cols = ['year', 'odometer']
cat_cols = [col for col in train_df.columns if col not in non_cols + num_cols]
```

```python
train_df = creansing_dataset(train_df)
test_df = creansing_dataset(test_df)
```

```python
def check_data_bias(comb_n):
    agg_l = []

    comb_list = itertools.combinations(cat_cols, comb_n)
    for comb in tqdm(comb_list):
        comb = list(comb)
        gcomb = comb + ['id']
        info_df = \
            pd.concat([
                train_df[gcomb].groupby(comb)['id'].count().rename('train'), 
                test_df[gcomb].groupby(comb)['id'].count().rename('test')
            ], axis=1).reset_index()

        agg_l.append([
            '-'.join(comb),
            info_df[info_df['test'].isnull()]['train'].sum(),
            info_df[info_df['train'].isnull()]['test'].sum()
        ])

    return pd.DataFrame(agg_l, columns=['comb', 'only_train', 'only_test'])
```

```python
comb2_df = check_data_bias(2)
display(comb2_df.sort_values('only_train', ascending=False).head(10))
display(comb2_df.sort_values('only_test', ascending=False).head(10))
```
